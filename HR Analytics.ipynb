```python
df.columns
```

```python
df['AttritionFlag'] = df['Attrition'].apply(lambda x: 1 if x=='Yes' else 0)
```

```python
df['IncomeBand'] = pd.cut(df['MonthlyIncome'], bins=[0,5000,10000,20000], labels=['Low','Mid','High'])
```

```python
df['CompositeSatisfaction'] = df[['JobSatisfaction','EnvironmentSatisfaction','RelationshipSatisfaction','WorkLifeBalance']].mean(axis=1)
```

```python
labels=["Near","Managable","Far"]
df["DistanceFromHome"]=pd.qcut(df["DistanceFromHome(in km)"],q=3,labels=labels)
```

```python
df[["EmployeeId","DistanceFromHome(in km)","DistanceFromHome"]].head(10)
```

```python
labels=["Young","Adult","Middle-Aged","Senior"]
df["AgeGroup"]=pd.qcut(df["Age"],q=4,labels=labels)
```
